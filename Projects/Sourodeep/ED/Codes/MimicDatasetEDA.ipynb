```python
import gzip
import pandas as pd
import sweetviz as sv
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
folder_path = '..\Data\mimic_iv_ed_2_2\ed'
current_path = os.getcwd()
os.chdir(folder_path)
```

```python

def get_all_files():
  files = []
  for file_name in os.listdir():
#     file_path = os.path.join(os.getcwd(), file_name)
    if os.path.isfile(file_name):
        if file_name.split('.')[-1] == 'gz':
            files.append(file_name)
  return files


files = get_all_files()

print(files)
dataframes = {}
for f in files:
    with gzip.open(f, 'rb') as f_in:
    # Read the contents of the .gz file
        csv_data = f_in.read()
        f_csv = f.split('.')[0]+'.csv'


    with open(f_csv, 'wt', encoding="utf-8") as f_out:
        # Write the contents of the CSV data to the new file
        f_out.write(csv_data.decode())

    variable_name = f.split('.')[0]
#     sweet_viz_eda(f_csv)
    df = pd.read_csv(f_csv)
    print(f_csv, df.shape)
    print(f)
    dataframes[variable_name] = df
    
    if f != 'edstays.csv.gz':
        os.remove(f.split('.')[0]+'.csv')
    display(df.head())
    
```

```python
def sweet_viz_eda(data, csv_file):
    
#     data = pd.read_csv(csv_file)
    report = sv.analyze(data)
    filename = 'report_'+csv_file+'.html'

    report.show_html('../../../Codes/Report/'+filename, layout='vertical')

```
