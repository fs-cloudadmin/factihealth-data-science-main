# Import Libraries

```python
import os
import pandas as pd
import sweetviz as sv
```

# Load Data
1. Admission
2. Diagnosis
3. EDStays

```python
admission_data = pd.read_csv('Data/admissions.csv')
diagnosis_data = pd.read_csv('Data/diagnosis.csv')
edstays_data = pd.read_csv('Data/edstays.csv')
vital_data = pd.read_csv('Data/vitalsign.csv')
print(admission_data.shape)
display(admission_data.head())
print(diagnosis_data.shape)
display(diagnosis_data.head())
print(edstays_data.shape)
display(edstays_data.head())
print(vital_data.shape)
display(vital_data.head())
```

```python
vital_data.sort_values(['subject_id', 'charttime'], inplace=True)
vital_data.head(29)
```

```python
vital_data.groupby(['subject_id', 'stay_id'])[['temperature','heartrate','resprate','o2sat','sbp','dbp']].mean()
```

# Map Dataset together

## Vitals and Diagnosis

```python
diag_vital_data = pd.merge(diagnosis_data, vital_data, how='left', on='stay_id')
print(diag_vital_data.shape)
diag_vital_data.tail(100)
```

```python
diag_edstay_data = pd.merge(diagnosis_data, edstays_data, how='left', on='stay_id')
print(diag_edstay_data.shape)
diag_edstay_data.tail(100)
```

```python
diag_edstay_data.columns
```

```python
merged_data = pd.merge(admission_data, diag_edstay_data, how='left', on='hadm_id')
merged_data.drop(['subject_id_x', 'subject_id_y', 'race_y'], inplace=True, axis=1)
merged_data.rename(columns={'race_x':'race'}, inplace=True)
print(merged_data.shape)
merged_data.tail(100)
```

```python
merged_data['admittime'] = pd.to_datetime(merged_data['admittime'])
merged_data['dischtime'] = pd.to_datetime(merged_data['dischtime'])
merged_data['admitdate'] = merged_data['admittime'].dt.date
merged_data.sort_values('admittime', inplace=True)
merged_data.tail(100)
```

```python
useful_cols = ['subject_id', 'hadm_id', 'stay_id', 'admittime', 'dischtime', 'admission_type', 'admission_location', 
               'discharge_location', 'insurance', 'marital_status', 'race', 'icd_title', 'gender', 'arrival_transport', 
               'disposition', 'admitdate']
```

```python
merged_data = merged_data[useful_cols]
```
