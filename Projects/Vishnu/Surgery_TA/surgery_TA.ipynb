```python
import pandas as pd
import random
from datetime import timedelta, datetime

average_surgery_times = {
    'Orthopedic': 120 - 80,  # Rough average duration in minutes
    'Gastrointestinal': 180 - 80,  # Rough average duration in minutes
    'Cardiac': 240 - 80,  # Rough average duration in minutes
    'Neurosurgery': 300 - 80,  # Rough average duration in minutes
    'General': 100 - 80,  # Rough average duration in minutes
}

average_cleaning_times = {
    'Orthopedic': 13,  # Rough average duration in minutes
    'Gastrointestinal': 15,  # Rough average duration in minutes
    'Cardiac': 18,  # Rough average duration in minutes
    'Neurosurgery': 23,  # Rough average duration in minutes
    'General': 10,  # Rough average duration in minutes
}

# Define surgery types and availability of resources
surgery_types = ['Orthopedic', 'Cardiac', 'Neurosurgery', 'General', 'Gastrointestinal']
availability_of_resources = ['Low', 'Normal', 'High']

# Random seed for reproducibility
random.seed(42)

# Create an empty list to store the data
data = []

# Generate data for the last 5 years
start_date = datetime.today() - timedelta(days=365 * 5)
end_date = datetime.today()

# Generate data for each day in the last 5 years
current_date = start_date
surgery_id = 10001  # Initialize the surgery ID
while current_date <= end_date:
    # Generate 5 surgeries per day
    for i in range(5):
        surgery_type = random.choice(surgery_types)
        patient_age = random.randint(20, 80)
        patient_bmi = round(random.uniform(20, 35), 1)
        preoperative_preparation = random.randint(10, 30)
        surgeon_experience = random.randint(5, 25)
        anesthesia_time = random.randint(20, 60)
        surgical_team_size = random.randint(2, 8)
        Cleaning_Staff = random.randint(2, 5)
        emergency = random.choice(['Yes', 'No'])
        availability_of_resource = random.choice(availability_of_resources)


        # Generate a random time between 08:00 and 20:00 for surgery start time
        surgery_start_time = current_date.replace(hour=8, minute=0) + timedelta(minutes=random.randint(0, 720))
        
        surgery_end_time = surgery_start_time + timedelta(minutes=average_surgery_times[surgery_type] +
                                                          (patient_age * 0.5) +
                                                          (surgeon_experience * 1.5) +
                                                          (anesthesia_time * 0.3))
        
        surgery_start_time =  float(f"{surgery_start_time.hour}.{surgery_start_time.minute}")

        surgery_end_time = float(f"{surgery_end_time.hour}.{surgery_end_time.minute}")


        # Apply patterns to the target variables based on important features
        surgery_turnaround_time = average_surgery_times[surgery_type] + (patient_age * 0.5) + (
                    surgeon_experience * 1.5) + (anesthesia_time * 0.3)

        # Assume a linear relationship between turnover time and surgical team size
        turnover_time = average_cleaning_times[surgery_type] + (surgical_team_size * 2) + (Cleaning_Staff * 2)

        data.append([current_date, surgery_id, surgery_type, patient_age, patient_bmi, preoperative_preparation,
                     surgeon_experience,
                     anesthesia_time, surgical_team_size, Cleaning_Staff, emergency,
                     surgery_start_time, surgery_end_time,
                     availability_of_resource, surgery_turnaround_time, turnover_time])

        surgery_id += 1  # Increment the surgery ID

    current_date += timedelta(days=1)

# Create a DataFrame from the generated data
columns = ['Date','Surgery_ID', 'Surgery_Type', 'Patient_Age', 'Patient_BMI', 'Preoperative_Preparation', 'Surgeon_Experience',
           'Anesthesia_Time', 'Surgical_Team_Size', 'Cleaning_Staff', 'Emergency',
           'Surgery_Start_Time', 'Surgery_End_Time', 'Availability_of_Resources', 'Surgery_Turnaround_Time',
           'Turnover_Time']

df = pd.DataFrame(data, columns=columns)

# Save the DataFrame to a CSV file
df.to_csv('../data/surgery_TA_dataset.csv', index=False)
```

```python
cat_coumns = ["Surgery_Type","Emergency","Availability_of_Resources"]

num_column =['Patient_Age', 'Patient_BMI','Preoperative_Preparation', 'Surgeon_Experience', 'Anesthesia_Time','Surgical_Team_Size', 'Cleaning_Staff','Surgery_Start_Time','Surgery_Turnaround_Time', 'Turnover_Time']

dummies = pd.get_dummies(data[cat_coumns],drop_first=True)

data_new = pd.concat([dummies,data[num_column]],axis=1)
```

```python
data_new.columns
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Splitting data into train test split

target,features = data_new[["Surgery_Turnaround_Time","Turnover_Time"]],data_new.drop(["Surgery_Turnaround_Time","Turnover_Time"],axis=1)

features = features.astype("int")

x_train, x_test, y_train, y_test = train_test_split(features, target)

model = XGBRegressor()

model.fit(x_train,y_train)

### Evaluating the model

from sklearn.metrics import mean_absolute_error, r2_score,mean_squared_error

predicted = model.predict(x_test)

print("R2 Score is : ",r2_score(predicted,y_test))

print("Mean Squared Error is : ",mean_squared_error(predicted,y_test))

print("Mean Absolute Error is : ",mean_absolute_error(predicted,y_test))

model.save_model(r"..\models\surgery_TA_model.model")
```

```python
def get_surgery_TA_sample():

    surgery_TA_sample = {'Surgery_Type_Gastrointestinal': [0],
    'Surgery_Type_General': [0],
    'Surgery_Type_Neurosurgery': [0],
    'Surgery_Type_Orthopedic': [0],
    'Emergency_Yes': [0],
    'Availability_of_Resources_Low': [0],
    'Availability_of_Resources_Normal': [0],
    'Patient_Age': [0],
    'Patient_BMI': [0],
    'Preoperative_Preparation': [0],
    'Surgeon_Experience': [0],
    'Anesthesia_Time': [0],
    'Surgical_Team_Size': [0],
    'Cleaning_Staff': [0],
    'Surgery_Start_Time': [0]}

    return surgery_TA_sample
```

```python
def prepare_surgery_TA_sample(scheduled_surgery_details):
    
    surgery_TA_sample = get_surgery_TA_sample()

    if "Surgery_Type_"+scheduled_surgery_details["Surgery_Type"] in surgery_TA_sample.keys():

        surgery_TA_sample["Surgery_Type_"+scheduled_surgery_details["Surgery_Type"]] = [1]

    if scheduled_surgery_details["Emergency"]=="Yes":

        surgery_TA_sample['Emergency_Yes']=[1]

    if "Availability_of_Resources_"+scheduled_surgery_details["Availability_of_Resources"] in surgery_TA_sample.keys():

        surgery_TA_sample["Availability_of_Resources_"+scheduled_surgery_details["Availability_of_Resources"]] = [1]

    for key in surgery_TA_sample.keys():

        if key in scheduled_surgery_details.keys():

            surgery_TA_sample[key] = scheduled_surgery_details[key]


    return pd.DataFrame(surgery_TA_sample)
```
